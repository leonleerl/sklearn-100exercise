# P079 异常值检测-局部异常因子算法LOF

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
```

```python
np.random.seed(42)
```

```python
df = pd.read_csv("./p079.csv")
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
df = pd.read_csv("p079.csv")

# 绘制散点图
plt.figure(figsize=(8, 5))
plt.scatter(df["x"], df["y"], alpha=0.6, edgecolors='k')

# 设置标题和坐标轴标签
plt.title("Scatter Plot of p079.csv Data")
plt.xlabel("x1")
plt.ylabel("x2")

# 显示图形
plt.show()
```

```python
lof = LocalOutlierFactor(n_neighbors=20)
```

```python
data = df.values
```

```python
data
```

```python
y_pred = lof.fit_predict(data)
```

```python
y_pred
```

```python
df["lof"] = y_pred
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 根据 `lof` 值筛选数据
df_positive = df[df["lof"] == 1]  # 颜色为蓝色
df_negative = df[df["lof"] == -1] # 颜色为红色

# 绘制散点图
plt.figure(figsize=(8, 5))
plt.scatter(df_positive["x"], df_positive["y"], color='blue', alpha=0.6, edgecolors='k', label='lof = 1')
plt.scatter(df_negative["x"], df_negative["y"], color='red', alpha=0.6, edgecolors='k', label='lof = -1')

# 设置标题和坐标轴标签
plt.title("Scatter Plot of p079.csv with LOF Labels")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()  # 添加图例

# 显示图形
plt.show()
```

# P080 异常值检测-LOF统计离群点分布

```python
df["lof"].value_counts()
```

# P081 异常值检测-孤立森林Isolation Forest

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
np.random.seed(42)
df = pd.read_csv("./p081.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["item_length"], df["item_width"], alpha=0.6, edgecolors='k')

# 设置标题和坐标轴标签
plt.title("Scatter Plot of p079.csv Data")
plt.xlabel("item_length")
plt.ylabel("item_width")

# 显示图形
plt.show()
```

```python
outlier = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)
```

```python
outlier.fit(df)
```

```python
df['outlier_flag'] = outlier.predict(df)
```

```python
df
```

```python

df_positive = df[df["outlier_flag"] == 1]  # 颜色为蓝色
df_negative = df[df["outlier_flag"] == -1] # 颜色为红色

# 绘制散点图
plt.figure(figsize=(8, 5))
plt.scatter(df_positive["item_length"], df_positive["item_width"], color='blue', alpha=0.6, edgecolors='k', label='outlier_flag = 1')
plt.scatter(df_negative["item_length"], df_negative["item_width"], color='red', alpha=0.6, edgecolors='k', label='outlier_flag = -1')

# 设置标题和坐标轴标签
plt.title("Scatter Plot of p081.csv with LOF Labels")
plt.xlabel("item_length")
plt.ylabel("item_width")
plt.legend()  # 添加图例

# 显示图形
plt.show()
```
