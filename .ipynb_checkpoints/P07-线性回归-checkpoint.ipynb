# P034 使用Numpy实现线性回归

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "years":[1,2,3,4,5,6],
    "salary":[4000,4250,4500,4750,5000,5250]
    }
)
df
```

```python
m = len(df)
m
```

```python
X1 = df["years"].values
X1
```

```python
X = np.append(np.ones((m,1)), X1.reshape(m, 1), axis=1)
X
```

```python
y = df["salary"]
```

```python
np.dot(X.T,y)
```

```python
coefs = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
coefs
```

```python
print("y={:.2f}+{:.2f}x".format(coefs[0], coefs[1]))
```

# P035 线性回归-sklearn实现线性回归

```python
from sklearn.linear_model import LinearRegression
```

```python
df
```

```python
model = LinearRegression()
```

```python
model.fit(
    df[["years"]],
    df[["salary"]]
)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
print("y={:.2f}+{:.2f}x".format(model.intercept_[0], model.coef_[0][0]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("./p036.csv")

# Prepare X (independent variable) and y (dependent variable)
X = df[['variable']].values  # Reshape to 2D array
y = df['target'].values

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Generate predictions for plotting the regression line
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)  # Generate 100 points in X range
y_pred = model.predict(X_range)

# Plot the data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.5, label="Data Points")  # Scatter plot

# Plot the regression line
plt.plot(X_range, y_pred, color='red', linewidth=2, label="Regression Line")  # Linear regression line

# Labels and title
plt.xlabel("Variable (X)")
plt.ylabel("Target (Y)")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
